### Задание 2 (необязательное)
Напишите функцию date_range, которая возвращает список дат за период от start_date до end_date. Даты должны вводиться в формате yyyy-mm-dd.
В случае неверного формата или при start_date > end_date должен возвращаться пустой список.

```python
from datetime import datetime, timedelta
```

```python
start_date = '2022-01-01'
end_date = '2022-01-05'
```

```python
# превращаем строчки в даты для дальнейшей работы с новыми переменными
start = datetime.strptime(start_date, '%Y-%m-%d')
end = datetime.strptime(end_date, '%Y-%m-%d')
```

```python
def date_range(start_date, end_date):
    try:
        start = datetime.strptime(start_date, '%Y-%m-%d')
        end = datetime.strptime(end_date, '%Y-%m-%d')
    except ValueError:
        # Если формат неверный (или дата некорректна) – то пустой список
        return []
    if start > end:
        # Если нач дата больше конечной – то пустой список
        return []
    else:
        # Иначе - нужно сгенерировать новый список с датами и добавить их
        list_date = []
        while start <= end:
            list_date.append(start.strftime('%Y-%m-%d'))
            start += timedelta(days=1)

        return list_date

# результат
print(date_range('2022-01-01', '2022-01-13'))
# Должно быть: ['2022-01-01', '2022-01-02', '2022-01-03']

print(date_range('2022-01-03', '2022-01-01'))   # start > end
# Должно быть: []

print(date_range('не дата', '2022-01-01'))      # неверный формат
# Должно быть: []
```
