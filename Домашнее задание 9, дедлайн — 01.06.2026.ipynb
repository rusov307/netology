```python
#создаем список нужных условий
pribaltika = ['Latvia', 'Lithuania', 'Estonia']
category = [4, 12, 21]
years = [2005, 2006, 2007, 2008, 2009, 2010]

#Фильтр по странам и категориям и годам
df_power[df_power['country'] == 'Estonia']  #and 'category' == category and 'year' == years)

#Фильтр по условиям из переменных
df_pribaltika = df_power.query('country in @pribaltika and category in @category and year in @years')
```
