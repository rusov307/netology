```python
#Создаём класс для получения курсов валют.
class Rate:
    # Метод __init__ - вызывается при создании объекта.
    def __init__(self):
        # В self.url храним ссылку на JSON с курсами ЦБ РФ.
        self.url = "https://www.cbr-xml-daily.ru/daily_json.js"


    def get_data_from_api(self):
        # GET-запрос к сайту и сохраняем ответ в переменную response.
        response = requests.get(self.url)

        # Проверяем ответ от ссылки.
        # Если будет 200, значит запрос выполнен успешно.
        if response.status_code == 200:
            # Дальше в словарь JSON.
            data = response.json()
            return data
        else:
            # Если ошибка подключения, то выводим сообщение и возвращаем пустой словарь.
            print("Ошибка при запросе к API. Код ответа:", response.status_code)
            return {}

    # Метод для поиска валюты с максимальным курсом.
    def currency_max_value(self):
        data = self.get_data_from_api()

        # Из структуры JSON забираем словарь с валютами, все валюты лежат внутри ключа 'Valute'.
        valute = data.get('Valute', {})

        # Если словарь пустой (например, если сервер не ответил), то возвращаем None и выходим из метода.
        if not valute:
            print("Не удалось получить список валют.")
            return None

        max_value = 0.0 # самое большое значение курса.
        max_name = "" # переменная для названия валюты с этим курсом.

        for char_code, info in valute.items(): # по .items() найти пары (код валюты, информация).
            # Из информации о валюте берём текущий курс (поле 'Value') и название валюты (поле 'Name').
            current_value = info.get('Value', 0.0)
            current_name = info.get('Name', '')

            # Сравниваем курс текущей валюты с максимальным найденным.
            # Если текущий курс больше, обновляем max_value и max_name.
            if current_value > max_value:
                max_value = current_value
                max_name = current_name

        print(f"Валюта с максимальным курсом: {max_name} ({max_value:.4f} руб.)")

        # Возвращаем название валюты по заданию.
        return max_name

 # Точка входа в программу.
# Ученик уже знает, что код лучше писать внутри проверки __name__.
if __name__ == "__main__":
    # Создаём объект нашего класса.
    checker = Rate()

    # Вызываем у объекта метод поиска валюты с максимальным курсом
    # и сохраняем результат в переменную result.
    result = checker.currency_max_value()

    # Дополнительно выводим результат в удобном формате.
    if result:
        print("Ответ:", result)
    else:
        print("Не удалось определить валюту с максимальным курсом.")
```
