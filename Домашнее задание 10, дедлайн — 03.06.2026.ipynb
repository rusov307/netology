### Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:

- оценка 2 и ниже — низкий рейтинг;
- оценка 4 и ниже — средний рейтинг;
- оценка 4.5 и 5 — высокий рейтинг.
##### Результат классификации запишите в столбец class.

```python
import pandas as pd
from tornado.gen import Return
```

```python
df_movies = pd.read_csv('C:\\Users\\User\\Desktop\\Аналитик данных\\ДЗ\\ДЗ 10 классы\\movies.csv', sep=',', encoding='utf-8')
```

```python
df_ratings = pd.read_csv('C:\\Users\\User\\Desktop\\Аналитик данных\\ДЗ\\ДЗ 10 классы\\ratings.csv', sep=',', encoding='utf-8')
```

```python
df_movies.head(5) #чтение файла с фильмами
```

```python
df_ratings.head(5)
```

```python
# пишу функцию для условий оценки
def class_rating(ocenka):
    '''
- оценка 2 и ниже — низкий рейтинг;
- оценка 4 и ниже — средний рейтинг;
- оценка 4.5 и 5 — высокий рейтинг.
    '''
    if ocenka <= 2.0:
        return 'низкий рейтинг'
    elif ocenka > 2.0 and ocenka <=4.0:
        return 'средний рейтинг'
    elif ocenka >= 4.5 and ocenka <=5.0:
        return 'высокий рейтинг'
    else:
        return 'между 4 и 4.5'
```

```python
# применяю функцию к файлу с рейтингами по столбцу рейтинг и создаю новый столбец класс
df_ratings['class'] = df_ratings['rating'].apply(class_rating)
df_ratings
```

```python
df_ratings[df_ratings['class'] == 'между 4 и 4.5']
```

Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

*** geo_data = {
‘Центр’: [‘москва’, ‘тула’, ‘ярославль’],

‘Дальний Восток’: [‘владивосток’, ‘сахалин’, ‘хабаровск’]
}

#### Результат классификации запишите в отдельный столбец region.

```python
df_keywords = pd.read_csv('C:\\Users\\User\\Desktop\\Аналитик данных\\ДЗ\\ДЗ 10 классы\\keywords.csv', sep=',', encoding='utf-8')
```

```python
cities = ['москва', 'тула', 'ярославль', 'петербург', 'псков', 'мурманск', 'владивосток', 'сахалин', 'хабаровск']
```

```python
centr = ['москва', 'тула', 'ярославль']
sev_zap = ['петербург', 'псков', 'мурманск']
dal_vost = ['владивосток', 'сахалин', 'хабаровск']
```

```python
# пишу функцию для принадлежности к региону
'''
geo_data = {
‘Центр’: [‘москва’, ‘тула’, ‘ярославль’],
‘Северо-Запад’: [‘петербург’, ‘псков’, ‘мурманск’],
‘Дальний Восток’: [‘владивосток’, ‘сахалин’, ‘хабаровск’]
}
'''
def geo_data(region):
    if region in centr:
        return 'Центр'
    elif region in sev_zap:
        return 'Северо-Запад'
    elif region in dal_vost:
        return 'Дальний Восток'
    else:
        return 'undefined'
```

```python
# применяю функцию к файлу keywords
df_keywords['region_row'] = df_keywords['keyword'].apply(geo_data)
df_keywords
```

```python
df_keywords[df_keywords['region_row'] == 'Северо-Запад']
```
